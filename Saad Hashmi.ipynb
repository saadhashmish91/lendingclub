# Part II - Cleaning and Manipulating the Data

### Let's start with reducing the number of columns since there are 100+

```python
#Identifying columns with only null values
columns_with_nulls = loan.columns[loan.isnull().all()]
```

```python
#Dropping columns with only NULL Values to make data more readable
loan = loan.drop(columns=columns_with_nulls)
```

```python
#Confirming columns with NA only are removed
columns_with_na = loan.columns[loan.isna().all()]
columns_with_na
```

```python
#removing other columns that will not have effect on analysis
columns_to_remove2 = ['id','member_id', 'url', 'desc', 'zip_code','policy_code', 'application_type', 'initial_list_status', 'desc']
loan = loan.drop(columns=columns_to_remove2)
```

```python
#dropping all columns columnms that have only one value
columns_to_remove3 = ['chargeoff_within_12_mths','collections_12_mths_ex_med','tax_liens']
loan = loan.drop(columns=columns_to_remove3)
```

```python
#removing months from term and % from int_rate and converting into numerical datatype
loan['int_rate'] = loan['int_rate'].str.replace('%', '').astype(float)
loan['term'] = loan['term'].str.replace('months', '').astype(int)
```

```python
loan.info()
```

```python
correlation_matrix = loan.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Greens', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#Removing some other columns:
#Total Payment is highly correlated with Total Principal, Payment Iv, Total Iterest
#Loan amount is highly correlated with funded amount and funded amount inv
#acc_now_delinq and delinq_amount are predominantly blank
columns_to_remove4 = ['funded_amnt','funded_amnt_inv','total_pymnt_inv','total_rec_prncp','total_rec_int','acc_now_delinq','delinq_amnt']
```

```python
loan = loan.drop(columns=columns_to_remove4)
```

```python
loan.info()
```

```python
#Seeing the number of uinque values in each column
unique_counts = loan.nunique()
print(unique_counts)
```

```python
#Removing columns that are beyond scope of EDA
#title - categorical variable has ~20K unique values. We will use purpose as an alternative for this
#emp_title - categorical variable has ~30K unique values 
#payment plan: only 1 value
#recoveries and collection_recovery_fee is 0.8 correlation
#out_prncp and out_prncp_inv are 100% correlated

columns_to_remove5 = ['title','emp_title','pymnt_plan']

loan = loan.drop(columns=columns_to_remove5)
```

```python
columns_to_remove6 = ['out_prncp_inv','collection_recovery_fee']
loan = loan.drop(columns=columns_to_remove6)
```

```python
correlation_matrix = loan.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Greens', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

### Managing NULL values

```python
loan.isnull().sum()
```

```python
#removing columns that have too many NULL values. Also, pub_rec and pub_rec_bankruptcy have high correlation (0.85)
columns_to_remove7 = ['mths_since_last_delinq','mths_since_last_record','next_pymnt_d','pub_rec_bankruptcies']
loan = loan.drop(columns=columns_to_remove7)
```

```python
#Focussing next of revol_util - converting into float and adjusting NULL values
loan['revol_util'] = loan['revol_util'].str.replace('%', '').astype(float)
loan['revol_util'] = loan['revol_util'].fillna(loan['revol_util'].median())
```

```python
loan.emp_length.value_counts()
```

```python
#Don't want to bias the employee length by filling with most common value
loan.emp_length = loan.emp_length.fillna('Not Available')
```

```python
#Changing dates from object to date datatype - issue_d, last_pymnt_d, last_credit_pull_d
loan['issue_d'] = pd.to_datetime(loan['issue_d'], format='%b-%y')
```

```python
loan['last_pymnt_d'] = pd.to_datetime(loan['last_pymnt_d'], format='%b-%y')
loan['last_credit_pull_d'] = pd.to_datetime(loan['last_credit_pull_d'], format='%b-%y')
```

```python
loan.isnull().sum()
```

### For some analysis, converting loan_status into integer type (as a new column) 
Will also delete current loan_status

```python
# Delete rows where 'loan_status' is 'Current'
loan = loan[loan['loan_status'] != 'Current']

loan['loan_status_convtd'] = loan['loan_status'].replace({'Charged Off': 1, 'Fully Paid': 0})

# Convert data type to int
loan['loan_status_convtd'] = loan['loan_status_convtd'].astype(int)
```

```python
#Let's see the spread of grades across different loan_status - Defaulters, current and paid

# Calculate the count of each combination of grades and loan status
counts = loan.groupby(['grade', 'loan_status']).size().reset_index(name='count')

total_count = counts['count'].sum()

# Creating pivot table: Grades are rows, loan_status as columns, and % of total entries as values
pivot = pd.pivot_table(counts, values='count', index='grade', columns='loan_status', aggfunc=lambda x: (x/total_count)*100)

sns.heatmap(pivot, cmap='Greens', annot=True)

plt.title('Weighted Average for Grades')
plt.show()
```

```python
#Since the number of loans are NOT equally distributed across different grades, above map doesn't show a true picture
#That's why I'm going to take a weighted average - across each category this time


# Calculate the weighted average across each row
row_weights_grades = pivot.sum(axis=1)
weighted_average_grades = (pivot.T / row_weights_grades).T

# Plot the heatmap using seaborn
sns.heatmap(weighted_average_grades, cmap='Greens', annot=True)

plt.title('Weighted Average for Grades')
plt.show()
```

#### Purpose seem to have a high impact on the default rate
Grade A default 6% of the time and the number gradually increases to 31% defaulters for F and G

```python
#Let's see the spread across different loan_status - with weighted average across category rows

# Calculate the count of each combination of grades and loan status
counts = loan.groupby(['purpose', 'loan_status']).size().reset_index(name='count')

total_count = counts['count'].sum()

# Creating pivot table: Grades are rows, loan_status as columns, and % of total entries as values
pivot = pd.pivot_table(counts, values='count', index='purpose', columns='loan_status', aggfunc=lambda x: (x/total_count)*100)

# Calculate the weighted average across each row
row_weights_purpose = pivot.sum(axis=1)
weighted_average_purpose = (pivot.T / row_weights_purpose).T

# Plot the heatmap using seaborn
sns.heatmap(weighted_average_purpose, cmap='Greens', annot=True)

plt.title('Weighted Average for Purpose')
plt.show()
```

Small Business loans have maximum default
Other reasons fall in between
